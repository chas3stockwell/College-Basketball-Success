The os functions allow the files to be iterated through easily in a folder, given a directory. As all files are found in the file data_torvik, all files except 2000 are iterated through and imported into a temporary dataframe structure. Each is then labeled with its corresponding year and then concatenated to the main dataframe.

```python
import os

directory_in_str = "./data_NBA"
directory = os.fsencode(directory_in_str)
yNum = 2001

for file in os.listdir(directory):
     filename = os.fsdecode(file)
     if filename.endswith(".csv") & ~("2000" in filename):
        df_temp = pd.read_csv((directory_in_str +"/"+ filename), sep=",", encoding='latin-1')
        df_temp["Year"] = yNum
        df_NBA = pd.concat([df_NBA, df_temp], ignore_index = True)
        yNum = yNum+1
     else:
         continue

df_NBA
```

One main discussion in looking at the meta of the NBA is the rise of the three point shot. With the rise of Stephen Curry came with it the change of offensive style of the game. Three point shots became known as the effecient shot to take. Valued at 50% more points than a shot made in the paint, it gives overall more points per possession given a decent three-point shot percentage. The increase in three-point shots have also increased spacing on the floor, increasing the availability for players in the post to make plays.

To quantify the prevalence of three point shots attempted, a new column is created for each player 3GA/(3GA+FGA)*100 to find the percentage of shots that were taken were three point shots.

```python
df_NBA["3P%_ofTotal"] = df_NBA["3PA"]/(df_NBA["3PA"]+df_NBA["FGA"])
df_NBA.head()
```

In order to better comprehend the change in 3 point attempts, a bar graph will suffice as the plot of choice in explaining the shift in the NBA scoring meta.

```python
#Grouping by each year, the percent of shots taken that were threes were averaged and then plotted as a bar per year.
attempts3_plot = df_NBA.groupby("Year")["3P%_ofTotal"].mean().plot.bar(figsize = (20, 10))
attempts3_plot.set_title("Percent of total shot attempts that were 3P attempts over Time")
attempts3_plot.set_ylabel("Percent of total shot attempts that were 3P attempts")
attempts3_plot.set_xlabel("Year")
attempts3_plot
```

As expected, the percentage of three point shots have increased over time as the culture has changed. While a significant increase in 3P attempts can be seen from 2010 and onwards, the consistent increase in 3P attempts can be seen 2012 and onwards, which is commonly known as the Golden Age or Dynasty of the Golden State Warriors.

However, let's look at the data more specifically as we have each individual data available. Does taking more 3 point attempts actually translate to more points per game?

```python
df_NBA["PPG"] = df_NBA["PTS"]/df_NBA["G"]
attempts3_scatter = df_NBA.plot.scatter(x = "3P%_ofTotal", y = "PPG", figsize = (20, 10), s = 0.10)
attempts3_scatter.set_title("Percent of total shot attempts that were 3P attempts vs. Average PPG")
attempts3_scatter.set_xlabel("Percent of total shot attempts that were 3P attempts")
attempts3_scatter.set_ylabel("Points per Game")
attempts3_scatter
```

Looking at the scatter plot, it can be seen that there are a significant number of players that never take a single three point attempt yet still attain a non-trivial number of points per game. You could say these were the Ben Simmons of the league.

```python
extra_drop = team_salaries[team_salaries['Salary'] == 'SALARY']

team_salaries = team_salaries.drop(extra_drop.index)

team_salaries['Player'] = team_salaries['Player'].str.split(',').str[0]

team_salaries
```

```python


extra = pd.merge(df_NBA, team_salaries, how = 'inner', on = ['Player', 'Year'])

extra
```

```python
extra['Salary'] = extra['Salary'].str[1:]

extra['Salary'] = extra['Salary'].str.replace("," , '')

#extra['Salary']

extra['Salary'] = extra['Salary'].astype(float)

#extra['Salary']

avg_sal = extra.groupby(["Year"])[["Salary"]].median()

print(avg_sal.head(15))
#avg_sal.head()


def player_wtscore_salary(salary, year, player):
    avg_score = avg_sal.loc[(year)]
    return salary/avg_score


extra["sal_score"] = extra.apply(lambda x: player_wtscore_salary(x["Salary"], x["Year"], x["Player"]), axis = 1)
extra.head()
```
